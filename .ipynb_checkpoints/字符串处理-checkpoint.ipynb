```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('C:/Users/Secretgarden/Desktop/工作簿1.csv',engine='python',encoding= 'utf-8',header=None)
```

```python
data = data[1:]
```

```python
data
```

**去空格**

```python
def deletet_blank_space(data):
    str_data = str(data)
    x = list()
    for i in str_data:
        if i.isalpha() or i == '·' or i == '/':
            x.append(i)
    return(x)
x = list()
for i in range(len(data)):
        x.append(deletet_blank_space(data[i:i+1]))
```

```python
len(x)
```

**去除n字符**

```python
def delete_n(data):
    delete_n = list()
    for i in data:
        if not i == 'n':
            delete_n.append(i)
    return(delete_n)
data_line = list()
for i in range(len(x)):
    data_line.append(delete_n(x[i]))
```

```python
names_line = list()
for i in range(len(data_line)):#把数据转化为字符串类型
    p = ''.join(data_line[i])
    names_line.append(p)
```

```python
names_line
```

**去除‘/’ 并且将名字提取到一个list中**

```python
names = list()
for i in range(len(names_line)):
    count = 0
    number = 0
    for j in range(len(names_line[i])):
        if names_line[i][j] == '/':
            count += 1
            if count <= 1:
                names.append(names_line[i][:j])
                number = j + 1
            else:
                names.append(names_line[i][number:j])
                number += j
        elif j == len(names_line[i]):
            names.append(names_line[i][count:j])
    
```

```python
names = list()
for i in range(5):
    count = 0
    number = 0
    for j in range(len(names_line[i])):
        if names_line[i][j] == '/':
            count += 1
            if count <= 1:
                names.append(names_line[i][:j])
                number = j + 1
            else:
                names.append(names_line[i][number:j])
                number += j
        elif j == len(names_line[i]):
            names.append(names_line[i][count:j])
    
```
